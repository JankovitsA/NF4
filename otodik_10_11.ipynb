```python
#adattiztítás
def process_prices(df):
  df['date']=df['date'].apply(lambda x: x[:10])
  df_filtered=df[['date','symbol','close']]
  #return df_filtered
  df_pivot=pd.pivot_table(df_filtered,#Pivot-táblába rendezzük
                        index=['date'],
                        columns=['symbol'],
                        values=['close'])
  df_pivot.columns=df_pivot.columns.droplevel()
  return df_pivot
```

```python
df_prices=process_prices(df_prices)

df_split=process_prices(df_split)
```

```python
print(df_prices.head(3))
```

```python
def add_effective_return(df): #effektív hozamot számolunk rá
  df_out=pd.DataFrame()
  for col in df.columns:
    df_out[col+'_effective_return']=df[col]/df[col].shift(1)-1
  return df_out
```

```python
df_prices_ret=add_effective_return(df_prices)
df_split_ret=add_effective_return(df_split) #a kiigazított adatsor
```

```python
print(df_prices_ret.shape,
df_split_ret.shape)
```

```python
df_merged=df_prices_ret.merge( #a normal és korrigált returnök összefésülése
    df_split_ret, on="date", suffixes=('_normal','_adjusted')
)
```

```python
df_merged.head(3)
```

```python
symbols=df_prices.columns
```

```python
splits=[]
for symbol in symbols:
  df_symbol=df_merged[[symbol+'_effective_return_normal',
                       symbol+'_effective_return_adjusted']]
  difference_array= \
              df_symbol[symbol+'_effective_return_normal'] - df_symbol[symbol+'_effective_return_adjusted']>0.001 #kivonjuk egymásból a hozamokat,
  if len(df_symbol.loc[difference_array])>0: #ha ezek közül a nem 0-k hossza 0-nál nagyobb (tehát van ilyen)
    splits.append(symbol) #akkor hozzáadjuk a számlálóhoz a nevét
print(splits,len(splits))
    #print(symbol, difference_array) 
```
